```python
import psutil
import pandas as pd
import time
from agent import generate_response  # Import the function
from utils import write_message

# Sample questions for testing
questions_1 = [
    "บริษัท ADVANC มีสินทรัพย์รวมในไตรมาสที่ 1 ปี 2019 เท่าไหร่",
    "บริษัท AOT มีหนี้สินรวมในไตรมาสที่ 1 ปี 2019 เท่าไหร่"
]

# Function to log query results
def log_query_results(question, response, chain_interaction, sql_query, result_data, chain_time, memory_usage):
    results.append({
        'Question': question,
        'Chain Interaction': chain_interaction,
        'SQL Query': sql_query,
        'Result Data': result_data,
        'Chain Time (s)': chain_time,
        'Response': response,
        'RAM Usage (MB)': memory_usage if memory_usage is not None else 0
    })

# Testing chatbot with detailed logging
def test_chatbot(questions):
    for question in questions:
        try:
            # Start RAM and timing metrics
            process = psutil.Process()
            ram_before = process.memory_info().rss / (1024 * 1024)  # RAM in MB
            start_time_chain = time.time()

            # Generate response
            result = generate_response(question)

            # Unpack results
            if isinstance(result, tuple) and len(result) == 4:
                response, chain_interaction, sql_query, result_data = result
            else:
                response, chain_interaction, sql_query, result_data = result, "Chain interaction not available", None, None

            # Calculate chain timing
            end_time_chain = time.time()
            chain_time = end_time_chain - start_time_chain

            # Measure RAM after execution
            ram_after = process.memory_info().rss / (1024 * 1024)  # RAM in MB
            memory_usage = ram_after - ram_before

            # Log the details
            log_query_results(question, response, chain_interaction, sql_query, result_data, chain_time, memory_usage)

        except Exception as e:
            # Log errors gracefully
            print(f"Error with question '{question}': {e}")
            log_query_results(question, f"Error: {e}", "Error in chain interaction", None, None, 0, None)

# Run tests
results = []
test_chatbot(questions_1)

# Save results to a CSV file
df_results = pd.DataFrame(results)
df_results.to_csv("chatbot_test_results.csv", index=False)
print("Test results have been saved to chatbot_test_results.csv.")
```
